```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Month': ['Jan','Feb','Mar','Apr','May','Jun'],
    'Sales_A': [220, 270, 300, 350, 400, 420],
    'Sales_B': [180, 230, 280, 310, 360, 390]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.plot(x='Month', y=['Sales_A', 'Sales_B'], kind='line', marker='o', linewidth=2)
```

```python
plt.title('Monthly Sales Comparison between Product A and B')
plt.xlabel('Month')
plt.ylabel('Sales (in Units)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Sales_A', 'Sales_B'], loc='upper left')
plt.tight_layout()
```

```python
plt.show()
```

```python
plt.style.use('seaborn-v0_8-darkgrid')

ax = df.plot(x='Month', y=['Sales_A', 'Sales_B'], marker='o', linewidth=3)
ax.set_title('Sales Performance - Styled Plot', fontsize=14, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (Units)', fontsize=12)
ax.legend(['Sales_A', 'Sales_B'], title='Product')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data ={
    'Month' : ['Jan','Feb','Mar','Apr','May','Jun'],
    'city_A' :[22,24,28,33,36,38],
    'city_B' :[25,27,29,32,34,37]
}
df_temp = pd.DataFrame(data)
print(df_temp)
```

```python
df_temp.plot(x='Month', y=['city_A','city_B'], kind='line', color=['blue','orange'])
```

```python
df_temp['diff'] = abs(df_temp['city_A'] - df_temp['city_B'])
```

```python
max_diff = df_temp['diff'].max()
idx = df_temp['diff'].idxmax()

month = df_temp.loc[idx, 'Month']
cityA = df_temp.loc[idx, 'city_A']
cityB = df_temp.loc[idx, 'city_B']
```

```python
import matplotlib.pyplot as plt

# Plot both city data
plt.plot(df_temp['Month'], df_temp['city_A'], marker='o', label='City A')
plt.plot(df_temp['Month'], df_temp['city_B'], marker='o', label='City B')

# Mark the max difference point
plt.scatter(month, cityA, color='red', s=100, label='Max Diff Point (City A)')
plt.scatter(month, cityB, color='red', s=100)

# Annotate
plt.text(month, (cityA + cityB) / 2,
         f"Max Diff = {max_diff}",
         ha='center', va='bottom', fontsize=10, color='red')

plt.xlabel("Month")
plt.ylabel("Temperature")
plt.title("Temperature Comparison & Max Difference Point")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

df_temp['diff'] = abs(df_temp['city_A'] - df_temp['city_B'])

plt.bar(df_temp['Month'], df_temp['diff'])
plt.title("Difference Between City A & City B")
plt.ylabel("Difference")
plt.xlabel("Month")

max_diff = df_temp['diff'].max()
plt.show()
```

```python
plt.plot(df_temp['Month'], df_temp['city_A'], marker='o', label='City A')
plt.plot(df_temp['Month'], df_temp['city_B'], marker='o', label='City B')

plt.fill_between(df_temp['Month'], df_temp['city_A'], df_temp['city_B'], alpha=0.2)

plt.title("Shaded Difference Area")
plt.legend()
plt.grid(True)
plt.show()
```

```python
df_temp['diff'] = abs(df_temp['city_A'] - df_temp['city_B'])
max_diff = df_temp['diff'].max()
idx = df_temp['diff'].idxmax()

month = df_temp.loc[idx, 'Month']
cityA = df_temp.loc[idx, 'city_A']
cityB = df_temp.loc[idx, 'city_B']

plt.plot(df_temp['Month'], df_temp['city_A'], marker='o', label='City A')
plt.plot(df_temp['Month'], df_temp['city_B'], marker='o', label='City B')

plt.scatter(month, cityA, s=150, color='red')
plt.scatter(month, cityB, s=150, color='red')

plt.text(month, (cityA + cityB)/2, f"Max Diff = {max_diff}",
         ha='center', va='bottom', fontsize=10, color='red')

plt.title("Maximum Difference Highlighted")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

x = np.arange(len(df_temp))

coefA = np.polyfit(x, df_temp['city_A'], 1)
coefB = np.polyfit(x, df_temp['city_B'], 1)

trendA = np.poly1d(coefA)(x)
trendB = np.poly1d(coefB)(x)

plt.plot(df_temp['Month'], df_temp['city_A'], 'o-', label='City A Data')
plt.plot(df_temp['Month'], df_temp['city_B'], 'o-', label='City B Data')
plt.plot(df_temp['Month'], trendA, '--', label='City A Trendline')
plt.plot(df_temp['Month'], trendB, '--', label='City B Trendline')

plt.title("Trendline Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
