```python
import pandas as pd

# Load your CSV file (example: if saved as employees.csv)
df = pd.read_csv("Practice Day.csv")

# Create Discounted_Price = Price - 10% discount
df["Discounted_Price"] = df["Price"] * 0.90   # or df["Price"] * (1 - 0.10)

# Save new file
df.to_csv("discounted_products.csv", index=False)

print("New file saved as discounted_products.csv")
```

```python
duplicate_counts = df.duplicated(subset='EmployeeID').sum()
print(f"Number of duplicate EmployeeID entries: {duplicate_counts}")
```

```python
# ---- 2. Remove duplicates ----
df_cleaned = df.drop_duplicates(subset='EmployeeID', keep='first')
print("before removing duplicates:", pd.read_csv("Practice Day.csv"))
print("after removing duplicates:", df_cleaned)
```

```python
df_missing = df_cleaned[(df_cleaned['Salary'].isnull()) | (df_cleaned['Age'].isnull())]

print("Rows with missing Salary or Age:")
print(df_missing)
```

```python
mean_age = df_cleaned['Age'].mean()        # calculate mean age
df_cleaned['Age'] = df_cleaned['Age'].fillna(mean_age)   # fill missing ages

print("Mean Age:", mean_age)
print(df_cleaned)
```

```python
df_cleaned = df_cleaned.dropna(subset=['Salary'])
print(df_cleaned)
```

```python
import numpy as np
max_marks1 = df['Marks1'].max()
max_marks2 = df['Marks2'].max() 
max_marks3 = df['Marks3'].max()
mean_marks1 = df['Marks1'].mean()
mean_marks2 = df['Marks2'].mean()
mean_marks3 = df['Marks3'].mean()
standard_deviation1 = np.std(df['Marks1'])
standard_deviation2 = np.std(df['Marks2'])
standard_deviation3 = np.std(df['Marks3'])
print("Maximum Marks in Subject 1:", max_marks1)
print("Maximum Marks in Subject 2:", max_marks2)
print("Maximum Marks in Subject 3:", max_marks3)
print("Mean Marks in Subject 1:", mean_marks1)
print("Mean Marks in Subject 2:", mean_marks2)
print("Mean Marks in Subject 3:", mean_marks3)
print("Standard Deviation in Subject 1:", standard_deviation1)
print("Standard Deviation in Subject 2:", standard_deviation2)
print("Standard Deviation in Subject 3:", standard_deviation3)
```

```python
mean_marks = df_cleaned[['Marks1', 'Marks2', 'Marks3']].mean()
max_marks  = df_cleaned[['Marks1', 'Marks2', 'Marks3']].max()

print("Mean of each subject:\n", mean_marks)
print("\nMaximum marks in each subject:\n", max_marks)
```

```python
df_cleaned['Total_Marks'] = df_cleaned['Marks1'] + df_cleaned['Marks2'] + df_cleaned['Marks3']

print("\nDataFrame with Total Marks:")
print(df_cleaned[['EmployeeID', 'EmployeeName', 'Total_Marks']])
```

```python
std_total = df_cleaned['Total_Marks'].std()

print("\nStandard Deviation of Total Marks:", std_total)
```

```python
import matplotlib.pyplot as plt

plt.hist(df_cleaned['Age'])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
plt.scatter(df_cleaned['Age'], df_cleaned['Salary'])
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary Scatter Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Make sure Discounted_Price exists
if 'Discounted_Price' not in df_cleaned.columns:
    df_cleaned['Discounted_Price'] = df_cleaned['Price'] * 0.90

# --------------------------
# 1️⃣ Histogram of Age
plt.figure(figsize=(6,4))
plt.hist(df_cleaned['Age'], bins=8, color='skyblue', edgecolor='black')
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()

# --------------------------
# 2️⃣ Scatter plot: Age vs Salary
plt.figure(figsize=(6,4))
plt.scatter(df_cleaned['Age'], df_cleaned['Salary'], color='green', s=80)
plt.title("Scatter Plot: Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()

# --------------------------
# 3️⃣ Scatter plot: Price vs Discounted Price
plt.figure(figsize=(6,4))
plt.scatter(df_cleaned['Price'], df_cleaned['Discounted_Price'], color='orange', s=80)
plt.title("Scatter Plot: Price vs Discounted Price")
plt.xlabel("Price")
plt.ylabel("Discounted Price")
plt.grid(True)
plt.show()

# --------------------------
# Optional 4️⃣ Bar plot: Salary of each employee
plt.figure(figsize=(8,4))
plt.bar(df_cleaned['EmployeeName'], df_cleaned['Salary'], color='purple')
plt.title("Salary of Employees")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.xticks(rotation=45)
plt.show()

# --------------------------
# Optional 5️⃣ Pie chart: Proportion of total Price per employee
plt.figure(figsize=(6,6))
plt.pie(df_cleaned['Price'], labels=df_cleaned['EmployeeName'], autopct='%1.1f%%', startangle=90)
plt.title("Price Contribution by Employee")
plt.show()
```
