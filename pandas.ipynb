iloc[row_start:row_end, column_start:column_end]

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'City': ['New York', 'London', 'Paris', 'Tokyo'],
    'Salary': [70000, 80000, 90000, 100000]
}, index=['a', 'b', 'c', 'd'])   # ← custom row labels

print(df)
```

```python
# 1. Select first 3 rows (all columns)
print(df.iloc[0:3])           # ← position based
print(df.loc['a':'c'])        # ← label based

# 2. Select first 3 columns (all rows)
print(df.iloc[:, 0:3])
print(df.loc[:, ['Name', 'Age', 'City']])

# 3. Select specific rows and columns
print(df.loc[['a','c','d'], ['Name','Salary']])   # by label
print(df.iloc[[0,2,3], [0,3]])                    # by position

# 4. Single cell
print(df.loc['b', 'City'])    # → 'London'
print(df.iloc[1, 2])          # → 'London'
```
